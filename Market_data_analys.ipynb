```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import networkx as nx
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

from google.colab import files
upload = files.upload()
```

```python
data=pd.read_csv("BreadBasket_DMS.csv")
data.head()
```

```python
data=data.dropna()
basket = data.groupby(['Transaction', 'Item'])['Item'].count().unstack().reset_index().fillna(0).set_index('Transaction')
basket = basket.applymap(lambda x: 1 if x > 0 else 0)
print(basket.head())
```

```python
frequent_itemsets = apriori(basket, min_support=0.01, use_colnames=True)
print(frequent_itemsets.head())
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.2)
print(rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']])
```

```python
import matplotlib.pyplot as plt

# Mendefinisikan ukuran titik berdasarkan skala lift
lift_scaled = (rules['lift'] - rules['lift'].min()) / (rules['lift'].max() - rules['lift'].min()) * 100 + 20

# Membuat scatter plot dengan matplotlib
plt.figure(figsize=(10, 6))

# Plot scatter antara support dan confidence, dengan ukuran titik berdasarkan lift
plt.scatter(rules['support'], rules['confidence'], s=lift_scaled, alpha=0.5, c=rules['lift'], cmap='viridis')

# Menambahkan label pada sumbu
plt.title('Scatter Plot of Association Rules: Support vs Confidence with Lift')
plt.xlabel('Support')
plt.ylabel('Confidence')

# Menambahkan color bar untuk skala lift
cbar = plt.colorbar()
cbar.set_label('Lift')

# Menampilkan plot
plt.show()
```
